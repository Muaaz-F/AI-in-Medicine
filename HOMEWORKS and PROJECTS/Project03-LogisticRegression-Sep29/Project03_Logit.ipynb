### Data Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import MinMaxScaler , StandardScaler
from sklearn.linear_model import LogisticRegression, Lasso, Ridge
from sklearn.metrics import confusion_matrix, classification_report
from sklearn import metrics
from sklearn.metrics import accuracy_score, recall_score, f1_score
```

```python
# Histograms and check the noisy data:

%matplotlib inline
fig, ax = plt.subplots(3,3,figsize=(15,9))
for i, col in enumerate(df):
    sns.histplot(df[col], kde=True, ax=ax[i//3, i%3])
plt.show()
```

```python
zero_col = ['Glucose','Insulin','SkinThickness','BloodPressure','BMI']
df[zero_col] = df[zero_col].replace(0, np.nan)
```

```python
df.isna().sum()
```

```python
for col in ['Glucose','Insulin','SkinThickness']:
    median_col = np.median(df[df[col].notna()][col])
    df[col] = df[col].fillna(median_col)
for col in ['BMI','BloodPressure']:
    mean_col = np.mean(df[df[col].notna()][col])
    df[col] = df[col].fillna(mean_col)
```

### Machine Learning

```python
x = df.drop('Outcome', axis=1)
y = df.Outcome
```

```python
scaler = StandardScaler()
x_norm = scaler.fit_transform(x)
x_norm
```

```python
x = pd.DataFrame(x_norm, index=x.index, columns=x.columns)
x.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
# Logistic Regression with L2 Regularization
l2_model = LogisticRegression(penalty='l2', C=1.0, max_iter=1000)
l2_model.fit(X_train, y_train)

# Predictions on the test set
y_pred_l2 = l2_model.predict(X_test)

# Accuracy Score
accuracy_l2 = accuracy_score(y_test, y_pred_l2)
print(f"L2 Regularization Accuracy: {accuracy_l2 * 100:.5f}%")
```

```python
# Logistic Regression with L1 Regularization
l1_model = LogisticRegression(penalty='l1', solver='liblinear', C=1.0, max_iter=1000)
l1_model.fit(X_train, y_train)

# Predictions on the test set
y_pred_l1 = l1_model.predict(X_test)

# Accuracy Score
accuracy_l1 = accuracy_score(y_test, y_pred_l1)
print(f"L1 Regularization Accuracy: {accuracy_l1 * 100:.5f}%")
```

```python
# Testing with different C values for L2 regularization

C_values = [0.01, 0.1, 1, 10, 100]
for c in C_values:
    l2_model = LogisticRegression(penalty='l2', C=c, max_iter=1000)
    l2_model.fit(X_train, y_train)
    y_pred_l2 = l2_model.predict(X_test)
    accuracy_l2 = accuracy_score(y_test, y_pred_l2)
    print(f"L2 Regularization with C={c}: {accuracy_l2 * 100:.5f}%")
```

### Final Model
